```python
import tensorflow as tf
import numpy as np

sample = " if you want you"
idx2char = list(set(sample))
char2idx = {c:i for i, c in enumerate(idx2char)}

dic_size = len(char2idx)
rnn_hidden_size = len(char2idx)
num_classes = len(char2idx)
batch_size = 1
sequence_length = len(sample) - 1

sample_idx = [char2idx[c] for c in sample]
x_data = [sample_idx[:-1]]
y_data = [sample_idx[1:]]

x_one_hot = tf.one_hot(x_data, num_classes)

cell = tf.keras.layers.LSTMCell(rnn_hidden_size)
lstm_layer = tf.keras.layers.RNN(cell, return_sequences=True)
loss_function = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True) # 제공된 입력값이 확률이 아닌 로짓임을 의미
                                                                                # 내부적으로 소프트맥스 함수를 적용하여 확률 분포 계산
optimizer = tf.keras.optimizers.Adam(learning_rate=0.1)

for i in range(2000):
    with tf.GradientTape() as tape:
        outputs = lstm_layer(x_one_hot, training=True)
        loss = loss_function(y_data, outputs)

    gradients = tape.gradient(loss, lstm_layer.trainable_variables)
    optimizer.apply_gradients(zip(gradients, lstm_layer.trainable_variables))

    if i % 100 == 0:
        prediction = tf.argmax(outputs, axis=2)
        print(f"{i} loss: {loss:.4f}, prediction: {prediction.numpy()}, true Y: {y_data}")
        prediction_str = [idx2char[c] for c in np.squeeze(prediction)]
        print("\tPrediction str:", "".join(prediction_str))
```

```python
import tensorflow as tf
import numpy as np

sentence = ("if you want to build a ship, don't drum up people together to "
            "collect wood and don't assign them tasks and work, but rather "
            "teach them to long for the endless immensity of the sea.")
char_set = list(set(sentence))
char_dic = {w:i for i, w in enumerate((char_set))}

dataX = []
dataY = []
sequence_length = 10
for i in range(0, len(sentence) - sequence_length):
    x_str = sentence[i:i+sequence_length]
    y_str = sentence[i+1:i+sequence_length+1]

    x = [char_dic[c] for c in x_str]
    y = [char_dic[c] for c in y_str]

    dataX.append(x)
    dataY.append(y)

data_dim = len(char_set)
hidden_size = len(char_set)
num_classes = len(char_set)
batch_size = len(dataX)

dataX_one_hot = tf.one_hot(dataX, num_classes)

cell = tf.keras.layers.LSTMCell(hidden_size)
cell = tf.keras.layers.StackedRNNCells([cell]*2)
lstm_layer = tf.keras.layers.RNN(cell, return_sequences=True)
loss_function = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True) # 제공된 입력값이 확률이 아닌 로짓임을 의미
                                                                                # 내부적으로 소프트맥스 함수를 적용하여 확률 분포 계산
softmax_w = tf.Variable(np.zeros((hidden_size, num_classes), dtype=np.float32))
softmax_b = tf.Variable(np.zeros(num_classes, dtype=np.float32))

optimizer = tf.keras.optimizers.Adam(learning_rate=0.1)

for i in range(500):
    with tf.GradientTape() as tape:
        outputs = lstm_layer(dataX_one_hot, training=True)
        X_for_softmax = tf.reshape(outputs, [-1, hidden_size])
        outputs = tf.matmul(X_for_softmax, softmax_w) + softmax_b
        outputs = tf.reshape(outputs, [batch_size, sequence_length, num_classes])
        loss = loss_function(dataY, outputs)

    gradients = tape.gradient(loss, lstm_layer.trainable_variables + [softmax_w, softmax_b])
    optimizer.apply_gradients(zip(gradients, lstm_layer.trainable_variables + [softmax_w, softmax_b]))

    for j, result in enumerate(outputs):
        index = np.argmax(result, axis=1)
        if j == 0:
            print("\t")
            print("".join([char_set[t] for t in index]), end="")
        else:
            print(char_set[index[-1]], end="")
```
