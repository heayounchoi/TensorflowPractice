```python
import tensorflow as tf

x_data = [[1., 2.],
          [2., 3.],
          [3., 1.],
          [4., 3.],
          [5., 3.],
          [6., 2.]]
y_data = [[0], [0], [0], [1], [1], [1]]

W = tf.Variable(tf.random.normal([2, 1]), name="weight")
b = tf.Variable(tf.random.normal([1]), name="bias")

optimizer = tf.optimizers.SGD(learning_rate=0.01)

def compute_cost(X, Y):
    hypothesis = tf.sigmoid(tf.matmul(X, W) + b)
    Y = tf.convert_to_tensor(Y, dtype=tf.float32)
    cost = -tf.reduce_mean(Y * tf.math.log(hypothesis) + (1 - Y) * tf.math.log(1 - hypothesis))
    return hypothesis, cost

def train_step(X, Y):
    with tf.GradientTape() as tape:
        hypothesis, cost = compute_cost(X, Y)
    gradients = tape.gradient(cost, [W, b])
    optimizer.apply_gradients(zip(gradients, [W, b]))
    return hypothesis, cost

for step in range(10001):
    hy_val, cost_val = train_step(x_data, y_data)
    predicted = tf.cast(hy_val > 0.5, dtype=tf.float32)
    accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, y_data), dtype=tf.float32))
    if step % 200 == 0:
        print(step, "Cost: ", cost_val.numpy(), "\nPredicted:\n", predicted.numpy(), "\nAccuracy: ", accuracy.numpy())
```
